Importing the Dependencies

```
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from google.colab import drive
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
drive.mount('/content/drive', force_remount=True)
```

Data Collection and Data Processing

```
file_path = '/content/drive/MyDrive/Copy of sonar data.csv'
sonar_data = pd.read_csv(file_path)
```

```
sonar_data.head()
```

```
sonar_data.shape
```

```
sonar_data.describe()
```

```
sonar_data.iloc[:, -1].value_counts()
```

M --> Mine

R --> Rock

```
last_column = sonar_data.columns[-1]
sonar_data.groupby(last_column).mean()
```

```
X = sonar_data.iloc[:, :-1]
Y = sonar_data.iloc[:, -1]
```

```
print(X)
print(Y)
```

Training and Test data

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```
print(X.shape, X_train.shape, X_test.shape)
```

```
print(X_train)
print(Y_train)
```

Model Training --> Logistic Regression

```
model = LogisticRegression()
```

```
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
conf_matrix = confusion_matrix(X_train_prediction, Y_train)
```

```
print('TRAINING DATA: ')
print('\nAccuracy on training data : ', training_data_accuracy)
print('\nConfusion Matrix : \n', conf_matrix)
```

```
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
conf_matrix1 = confusion_matrix(X_test_prediction, Y_test)
```

```
print('TEST DATA: ')
print('\nAccuracy on test data : ', test_data_accuracy)
print('\nConfusion Matrix : \n', conf_matrix1)
```

```
class_report = classification_report(Y_test, X_test_prediction)
print("Classification Report:")
print(class_report)
```

Analysis and Visulaising

```
# 1. Feature Distribution Histograms
plt.figure(figsize=(15, 20))
for i, col in enumerate(X.columns[:10]):  # Visualize first 10 features for brevity
    plt.subplot(5, 2, i + 1)
    sns.histplot(X[col], kde=True, color='skyblue')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```
# 2. Correlation Heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = X.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```
# 3. Confusion Matrix
conf_matrix = confusion_matrix(Y_test, X_test_prediction)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=model.classes_)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```
# 4. Accuracy Comparison
plt.figure(figsize=(6, 4))
accuracies = [training_data_accuracy, test_data_accuracy]
labels = ['Training Accuracy', 'Test Accuracy']
plt.bar(labels, accuracies, color=['blue', 'green'])
plt.ylim(0, 1)
plt.title('Training vs Test Accuracy')
plt.ylabel('Accuracy')
plt.show()
```
